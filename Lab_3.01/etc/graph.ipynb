```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
phi1 = np.array([1.71, 3.71, 5.71, 7.71, 9.71, 11.71, 12.66])
x1 = np.array([2.8, 6.6, 10.8, 15.5, 20.4, 25, 27.2])

phi2 = np.array([2.17, 4.17, 5.17, 6.17, 8.17, 9.17, 11.17, 12.29])
x2 = np.array([2.5, 5, 6.3, 7.3, 21.1, 22.6, 25.6, 27.4])

plt.plot(x1, phi1, 'ro', color='red', linestyle='--')
plt.plot(x2, phi2, 'ro', color='blue', linestyle='--')

plt.xlabel('X (см)')
plt.ylabel('Значение потенциала (В)')

for i_x, i_y in zip(x1, phi1):
    plt.text(i_x, i_y, i_y)

for i_x, i_y in zip(x2, phi2):
    plt.text(i_x, i_y, i_y)

plt.grid()

plt.savefig('main.png')
```
